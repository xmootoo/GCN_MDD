```python
# Load Data
import mne
import numpy as np

H_EO_data = []
H_EC_data = []
H_TASK_data = []
MDD_EO_data = []
MDD_EC_data = []
MDD_TASK_data = []

for i in range(len(file_names)):
  for j in range(len(file_names[i])):
    if i == 0:
      file =  "C:/Users/xmoot/Desktop/Jupyter/Mathematical Biology/Final Project/HUMS Dataset/H/EO/" + file_names[i][j]
      data = mne.io.read_raw_edf(file)
      H_EO_data.append(data.get_data())
    if i == 1:
      file = "C:/Users/xmoot/Desktop/Jupyter/Mathematical Biology/Final Project/HUMS Dataset/H/EC/" + file_names[i][j]
      data = mne.io.read_raw_edf(file)
      H_EC_data.append(data.get_data())
    if i == 2:
      file = "C:/Users/xmoot/Desktop/Jupyter/Mathematical Biology/Final Project/HUMS Dataset/H/TASK/" + file_names[i][j]
      data = mne.io.read_raw_edf(file)
      H_TASK_data.append(data.get_data())
    if i == 3:
      file = "C:/Users/xmoot/Desktop/Jupyter/Mathematical Biology/Final Project/HUMS Dataset/MDD/EO/" + file_names[i][j]
      data = mne.io.read_raw_edf(file)
      MDD_EO_data.append(data.get_data())
    if i == 4:
      file = "C:/Users/xmoot/Desktop/Jupyter/Mathematical Biology/Final Project/HUMS Dataset/MDD/EC/" + file_names[i][j]
      data = mne.io.read_raw_edf(file)
      MDD_EC_data.append(data.get_data())
    if i == 5:
      file = "C:/Users/xmoot/Desktop/Jupyter/Mathematical Biology/Final Project/HUMS Dataset/MDD/TASK/" + file_names[i][j]
      data = mne.io.read_raw_edf(file)
      MDD_TASK_data.append(data.get_data())

# Metadata included in the file and a list of all channels:
# info = data.info
# channels = data.ch_names
```

```python
# We will train on the TASK data (ignoring EC and EO data for now)
# Each EEG signal is varying in time length. We first find the EEG signal with least time
M_1 = min([np.shape(H_TASK_data[i])[1] for i in range(len(H_TASK_data))])
M_2 = min([np.shape(MDD_TASK_data[i])[1] for i in range(len(MDD_TASK_data))])

# Round down to nearest order of 10000
M = min(M_1, M_2) - 3856

# Cut the end of each EEG signal down to this minimum time length M
# Split into signal lengths of 10000 discrete time points each
# Take the average across each time point
for i in range(len(H_TASK_data)):
  length = np.shape(H_TASK_data[i])[1]
  if length > M:
    H_TASK_data[i] = np.delete(H_TASK_data[i], list(range(M, length)), 1)
    H_TASK_data[i] = np.split(H_TASK_data[i], 15, axis = 1)

for i in range(len(MDD_TASK_data)):
  length = np.shape(MDD_TASK_data[i])[1]
  if length > M:
    MDD_TASK_data[i] = np.delete(MDD_TASK_data[i], list(range(M, length)), 1)
    MDD_TASK_data[i] = np.split(MDD_TASK_data[i], 15, axis = 1)
```

```python
# Dimensionality reduction
# Since there are too many time points, we must reduce the data in meaningful manner to construct our features
# We begin by averaging the data across each 10000 discrete time points 
H_TASK_avg = np.zeros((28,15,22))
MDD_TASK_avg = np.zeros((33,15,22))

for i in range(len(H_TASK_data)):
    for j in range(15):
        for k in range(22):
            H_TASK_avg[i][j][k] = np.mean(H_TASK_data[i][j][k])

for i in range(len(MDD_TASK_data)):
    for j in range(15):
        for k in range(22):
            MDD_TASK_avg[i][j][k] = np.mean(MDD_TASK_data[i][j][k])
```

```python
# Compile our H_TASK_data samples and MDD_TASK_data samples into an array
# 0 corresponds to Healthy Control and 1 corresponds to MDD
total_TASK_data = [[(H_TASK_avg[i].flatten()).reshape(330,1), 0] for i in range(len(H_TASK_data))]
for i in range(len(MDD_TASK_data)):
  total_TASK_data.append([(MDD_TASK_avg[i].flatten()).reshape(330,1), 1])

# Shuffle our samples, otherwise this will interfere with training
import random
random.shuffle(total_TASK_data)
```

```python
# We now split the data into our input array X, and output array Y
# We also to flatten each of our inputs from total_TASK_data into vectors, so we get the input matrix X = [x^(1) x^(2) . . . x^(m)]
# We have n = 330 features per input vector, so our matrix will be of size 330 x 61
n = 330

input_vectors_TASK = [total_TASK_data[i][0] for i in range(len(total_TASK_data))]
X_matrix = np.dstack(input_vectors_TASK).squeeze()

output_vectors_TASK = [total_TASK_data[i][1] for i in range(len(total_TASK_data))]
Y_vector = np.dstack(output_vectors_TASK).squeeze()
Y_vector = Y_vector.reshape(-1,1)
```

```python
# Load data
import torch

X = torch.tensor(X_matrix, dtype = torch.float32)
Y = torch.tensor(Y_vector, dtype = torch.float32)
```

```python
# Model (Three layer neural network)
import torch.nn as nn

# Select layer sizes
# We want our layer sizes to be declining until we reach the final layer size of 1
n_0 = n
n_1 = int(n_0 * (2/3))
n_2 = int(n_1 * (4/5))
n_3 = int(n_2 * (8/10))
n_4 = int(n_3 * (4/5))
n_5 = int(n_4 * (3/5))
n_6 = int(n_5 * (2/5))
n_7 = int(n_6 * (2/5))

class three_layer_nn(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(n_0, n_1)
        self.activation1 = nn.LeakyReLU()
        self.layer2 = nn.Linear(n_1, n_2)
        self.activation2 = nn.LeakyReLU()
        self.layer3 = nn.Linear(n_2, n_3)
        self.activation3 = nn.LeakyReLU()
        self.layer4 = nn.Linear(n_3, n_4)
        self.activation4 = nn.LeakyReLU()
        self.layer5 = nn.Linear(n_4, n_5)
        self.activation5 = nn.LeakyReLU()
        self.layer6 = nn.Linear(n_5, n_6)
        self.activation6 = nn.LeakyReLU()
        self.layer7 = nn.Linear(n_6, n_7)
        self.activation7 = nn.LeakyReLU()
        self.output = nn.Linear(n_7, 1)
        self.sigmoid = nn.Sigmoid()
 
    def forward(self, x):
        x = self.activation1(self.layer1(x))
        x = self.activation2(self.layer2(x))
        x = self.activation3(self.layer3(x))
        x = self.activation4(self.layer4(x))
        x = self.activation5(self.layer5(x))
        x = self.activation6(self.layer6(x))
        x = self.activation7(self.layer7(x))
        x = self.sigmoid(self.output(x))
        return x
```

```python
# Learning procedure

import copy
import torch.optim as optim
import tqdm

def training(model, X_train, Y_train, X_test, Y_test):
  # Select the loss function (cross entropy)
  loss_funct = nn.BCELoss()
  
  # Adams optimizier: stochastic gradient descent (SGD), for our learning procedure
  optimizer = optim.Adam(model.parameters(), lr = 0.0001)
  
  # Number of iterations
  epochs = 500

  # Batch size for SGD
  batch_size = 10
  batch_start = torch.arange(0, len(X_train), batch_size)

  best_acc = - np.inf
  optimal_weights = None

  for epoch in range(epochs):
    model.train()
    with tqdm.tqdm(batch_start, unit = 'batch', miniinterval = 0, disable = True) as bar:
      bar.set_description(f"Epoch {epoch}")
      for start in bar:
          # Select a batch
          X_batch = X_train[start:start+batch_size]
          Y_batch = Y_train[start:start+batch_size]
          
          # Compute the forward pass
          Y_hat = model(X_batch)
          loss = loss_funct(Y_hat, Y_batch)
          
          # Compute the backwards pass
          optimizer.zero_grad()
          loss.backward()
          
          # Perform weight update
          optimizer.step()
          
          # Print progress
          acc = (Y_hat.round() == Y_batch).float().mean()
          bar.set_postfix(
            loss=float(loss),
            acc=float(acc)
          )

    # At the end of each iteration, evaluate the model accuracy
    model.eval()
    Y_hat = model(X_test)
    acc = (Y_hat.round() == Y_test).float().mean()
    acc = float(acc)
    if acc > best_acc:
      best_acc = acc
      optimal_weights = copy.deepcopy(model.state_dict())
      
      # Restore the model and return the highest accuracy
  model.load_state_dict(optimal_weights)
  return best_acc
```

```python
# Evaluate model performance

from sklearn.model_selection import StratifiedKFold, train_test_split

# train-test split: Hold out the test set for final model evaluation
X_train, X_test, Y_train, Y_test = train_test_split(X.T, Y, train_size=0.7, shuffle=True)

# We use 5-fold Cross Validation (CV) to determine our accuracy
kfold = StratifiedKFold(n_splits=10, shuffle=True)
cv_scores = []
for train, test in kfold.split(X_train, Y_train):
    # Create the model, train on the data, then return model accuracy
    model = three_layer_nn()
    acc = training(model, X_train[train], Y_train[train], X_train[test], Y_train[test])
    print("Accuracy: %.2f" % acc)
    cv_scores.append(acc)
 
# Model evaluation
acc = np.mean(cv_scores)
std = np.std(cv_scores)
print("Deep: %.2f%% (+/- %.2f%%)" % (acc*100, std*100))
```
